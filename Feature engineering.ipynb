My plan is to perform sentiment analysis on the title and description columns of the data. In order to be able to effectively use the textual columns, I will need to clean them.

Here are the steps I am going to perform:

1. Expanding contractions of words (for example, changing "you're" to "you are."
2. Removing all the hyperlinks from the text (i.e. links starting with "http").
3. Changing "&" to "and."
4. Removing the newline character ("\\n).
5. Removing all non-alphabet characters.
6. Removing unnecessary whitespace (by reducing it to just 1 space between words).
7. Removing stopwords (by using NLTK's stopwords list).

Before we commence the cleaning process, let us look at what a video description looks like now so that we can compare it to a cleaned description.

### Cleaning the 'title' and 'description' columns for sentiment analysis

#### Expanding contractions

```python
def decontracted(phrase):
    # specific
    phrase = phrase.lower()
    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    phrase = re.sub(r"y\'all", "you all", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase

df['description'] = df['description'].apply(decontracted)
df['title'] = df['title'].apply(decontracted)
```

#### Removing hyperlinks, emojis, and all non-alphabet characters

```python
def cleaning(text):
    text = re.sub('http[s]?://\S+', '', text)
    text = re.sub("&", 'and', text)
    text = text.replace('\\n', '')
    text = re.sub("[^a-zA-Z]"," ",text)
    return text.lower()

df["description"]= df["description"].apply(cleaning)
df["title"] = df["title"].apply(cleaning)
```

```python
porter = nltk.stem.PorterStemmer()

def stemming(sentence):
    token_words = nltk.tokenize.word_tokenize(sentence)
    stem_sentence = []
    for word in token_words:
        stem_sentence.append(porter.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)

df["description"]= df["description"].apply(stemming)
df["title"] = df["title"].apply(stemming)
```

```python
def final_clean(text):
    text = re.sub(r"\s\s+" , " ", text)
    return text

df["description"]= df["description"].apply(final_clean)
df["title"] = df["title"].apply(final_clean)
```

Yes! We have successfully cleaned up the textual columns.

#### Removing stopwords

```python
stopword_list = nltk.corpus.stopwords.words('english')
```

```python
df['description'] = df['description'].apply(lambda x: ' '.join([item for item in x.split() if item not in stopword_list]))
df['title'] = df['title'].apply(lambda x: ' '.join([item for item in x.split() if item not in stopword_list]))
```
